```python
import pandas as pd

df = pd.read_excel('linechart.xlsx')
df.head()
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(12,6))
plt.plot(df['Quarter'], df['Fridge'], color='blue', label='Fridge')
plt.plot(df['Quarter'], df['Dishwasher'],color='orange')
plt.plot(df['Quarter'], df['Washing Machine'],color='teal')
plt.title('Product sales')
plt.ylabel('Revenue')
plt.xlabel('Quarter')
plt.legend()
plt.show()
```

```python
sum2 = df[['Fridge', 'Dishwasher','Washing Machine']].sum()
sum2
```

```python
sum2.index
```

```python
plt.pie(sum2, labels=sum2.index, autopct= '%1.2f%%', explode=(0.1,0.2,0.05),shadow=True, startangle= 140)
plt.show()
```
