```python
import pandas as pd
df = pd.read_csv('weather_data.csv', parse_dates=['day'])
df
```

```python
type(df.day[0])
```

```python
df.set_index('day',inplace=True)
df
```

```python
ds= df.fillna(0)
ds
```

```python
## Always use inplace if altering the same data frame
ds = df.fillna({'temperature':df.temperature.mean(), 'windspeed':df.windspeed.mean(),})
ds
```

```python
## ffill add the value from the previous row - forward filling
ds = df.fillna(method='ffill')
ds
```

```python

ds = df.fillna(method='bfill', axis ='columns')
ds
```
