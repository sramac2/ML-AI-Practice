```python
import csv
import pandas
def calculate_rating_stats(data, industry=None):
    valid_reviews =[]
    for row in data:
        if row[3]!='NULL' and (not industry or row[1]==industry):
            valid_reviews.append(float(row[3]))
    max_rating = max(valid_reviews)
    min_rating = min(valid_reviews)
    avg = sum(valid_reviews)/len(valid_reviews)
    return max_rating, min_rating, avg

with open("movies.csv") as m:
    data = list(csv.reader(m))
    print(calculate_rating_stats(data=data[1:]))
    print(calculate_rating_stats(data=data[1:],industry='Hollywood'))
    print(calculate_rating_stats(data=data[1:],industry='Bollywood'))

```
