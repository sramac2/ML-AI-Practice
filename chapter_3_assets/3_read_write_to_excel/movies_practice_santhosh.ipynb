```python
import pandas as pd

#second parameter is saying the sheet name
df_movies = pd.read_excel('movies_db.xlsx', 'movies')
type(df_movies)
df_movies.head()
```

```python
df_financials = pd.read_excel('movies_db.xlsx', 'financials')
type(df_financials)
df_financials.head()
```

```python
def standard_c(curr):
    if curr == '$$' or curr =='Dollars':
        return 'USD'
    else:
        return curr
df_financials = pd.read_excel('movies_db.xlsx', 'financials', converters={'currency':standard_c})
type(df_financials)
df_financials.head()
```

```python
df_merged = pd.merge(df_movies, df_financials, on ='movie_id')
df_merged.head()
```

```python
df_merged.to_excel('movies_merged.xlsx', sheet_name='movies',index=False)
```
